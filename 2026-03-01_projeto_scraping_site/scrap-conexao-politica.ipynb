```python
import requests
from bs4 import BeautifulSoup
```

```python
base_url1 = "https://www.conexaopolitica.com.br/analise/"
base_url2 = "https://www.conexaopolitica.com.br/opiniao/"
```

```python
html = requests.get(base_url1).content
soup = BeautifulSoup(html, 'html.parser')
```

```python
with open('conexao-politica-analise.html', 'w') as f:
    f.write(str(soup))
```

```python
from pydantic import BaseModel, Field

class Article(BaseModel):
    category: str = Field(..., description="The category of the article")
    title: str = Field(..., description="The title of the article")
    subtitle: str = Field(..., description="The subtitle of the article")
    url: str = Field(..., description="The URL of the article")
```

```python
articles = soup.find_all("article")
print(len(articles))
articles[:5]
```

```python
formatted_articles = []

for article in articles:
    category = article.find("p", class_="post-hat-feed-lg").text.strip()
    title = article.find("h2", class_="post-title-feed-lg").text.strip()
    subtitle = article.find("p", class_="feed-excert").text.strip()
    url = article.find("a")["href"]
    
    formatted_articles.append(Article(
        category=category,
        title=title,
        subtitle=subtitle,
        url=url
    ))

formatted_articles[:5]
```

## Getting all "Análise"

```python
import time

last_page = soup.find_all("a", class_="page-numbers")[-2]["href"]
last_page = int(last_page.split("/")[-2])
print(last_page)

formatted_articles = []

for page in range(1, last_page + 1):
    print(f"Scraping page {page} of {last_page}")
    html = requests.get(f"{base_url1}page/{page}/").content
    soup = BeautifulSoup(html, 'html.parser')
    
    articles = soup.find_all("article")
    
    for article in articles:
        category = article.find("p", class_="post-hat-feed-lg").text.strip()
        title = article.find("h2", class_="post-title-feed-lg").text.strip()
        subtitle = article.find("p", class_="feed-excert").text.strip()
        url = article.find("a")["href"]
        
        formatted_articles.append(Article(
            category=category,
            title=title,
            subtitle=subtitle,
            url=url
        ))
    time.sleep(1) 
```

```python
# safe save to jsonl
import json
with open('conexao-politica-analise-articles-links.jsonl', 'w') as f:
    for article in formatted_articles:
        f.write(json.dumps(article.model_dump()) + "\n")
```

```python
class ArticleContent(BaseModel):
    title: str = Field(..., description="The title of the article")
    category: str = Field(..., description="The category of the article")
    author: str = Field(..., description="The author of the article")
    date: str = Field(..., description="The publication date of the article")
    content: str = Field(..., description="The full content of the article")
    keywords: list[str] = Field(..., description="A list of keywords associated with the article")
    more_links: list[str] = Field(..., description="A list of additional links found in the article")
```

```python
from datetime import datetime

# parse this json: <script class="yoast-schema-graph" type="application/ld+json">


# Find and extract the JSON content from the script tag
script_tag = soup.find('script', class_='yoast-schema-graph')
if script_tag:
    json_content = script_tag.string
    try:
        parsed_json = json.loads(json_content)
        #print("Parsed JSON content:")
        #print(json.dumps(parsed_json, indent=2))
    except json.JSONDecodeError as e:
        print(f"Error parsing JSON: {e}")

title = parsed_json["@graph"][0]["headline"]
category = parsed_json["@graph"][0]["articleSection"][0]
author = parsed_json["@graph"][-1]["name"]
date = parsed_json["@graph"][0]["datePublished"]
content = parsed_json["@graph"][0]["articleBody"]
keywords = parsed_json["@graph"][0].get("keywords", [])
# a class feed-link
more_links = [a["href"] for a in soup.find_all("a", class_="feed-link")]

print(f"Title: {title}")
print(f"Category: {category}")
print(f"Author: {author}")
print(f"Date: {datetime.strptime(date, '%Y-%m-%dT%H:%M:%S%z').strftime('%Y-%m-%d %H:%M:%S')} (UTC-0)")
print(f"Content: {content[:200]}...")  # Print the first 200 characters of the content
print(f"Keywords: {keywords}")
print(f"More Links: {more_links}")
```

```python
def get_article_info(url: str) -> ArticleContent:
    article_content = requests.get(url).content
    soup = BeautifulSoup(article_content, 'html.parser')

    script_tag = soup.find('script', class_='yoast-schema-graph')
    if script_tag:
        json_content = script_tag.string
        try:
            parsed_json = json.loads(json_content)
        except json.JSONDecodeError as e:
            print(f"Error parsing JSON: {e}")
            return None

    title = parsed_json["@graph"][0]["headline"]
    category = parsed_json["@graph"][0]["articleSection"][0]
    author = parsed_json["@graph"][-1]["name"]
    date = parsed_json["@graph"][0]["datePublished"]
    content = parsed_json["@graph"][0]["articleBody"]
    keywords = parsed_json["@graph"][0].get("keywords", [])
    more_links = [a["href"] for a in soup.find_all("a", class_="feed-link")]

    return ArticleContent(
        title=title,
        category=category,
        author=author,
        date=datetime.strptime(date, '%Y-%m-%dT%H:%M:%S%z').strftime('%Y-%m-%d %H:%M:%S'),
        content=content,
        keywords=keywords,
        more_links=more_links
    )
```

## Getting all articles

```python
# from jsonls
articles_opiniao = []
articles_analise = []

with open('conexao-politica-opiniao-articles-links.jsonl', 'r') as f:
    for line in f:
        articles_opiniao.append(Article.model_validate_json(line))

with open('conexao-politica-analise-articles-links.jsonl', 'r') as f:
    for line in f:
        articles_analise.append(Article.model_validate_json(line))

print(f"Total opiniao articles: {len(articles_opiniao)}")
print(f"Total analise articles: {len(articles_analise)}")
```

```python
articles_opiniao_contents = []

print("=============== ARTIGOS OPINIÃO ===============")
for article in articles_opiniao:
    print(f"Getting content for article: {article.title}")
    article_info = get_article_info(article.url)
    if article_info:
        articles_opiniao_contents.append(article_info)
    time.sleep(1)  # Sleep to avoid overwhelming the server

with open('conexao-politica-opiniao-articles-content.jsonl', 'w') as f:
    for article in articles_opiniao_contents:
        f.write(json.dumps(article.model_dump()) + "\n")
```

```python

articles_analise_contents = []
print("=============== ARTIGOS ANÁLISE ===============")
for article in articles_analise:
    print(f"Getting content for article: {article.title}")
    article_info = get_article_info(article.url)
    if article_info:
        articles_analise_contents.append(article_info)
    time.sleep(1)  # Sleep to avoid overwhelming the server

with open('conexao-politica-analise-articles-content.jsonl', 'w') as f:
    for article in articles_analise_contents:
        f.write(json.dumps(article.model_dump()) + "\n")
```
