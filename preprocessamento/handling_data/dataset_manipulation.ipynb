## core

```python
import pandas as pd
from pathlib import Path
```

```python
labeled_df = pd.read_csv("datasets/labeled_df.csv")
labeled_df.head()
```

## limpeza

```python
labeled_df.isnull().sum()
```

```python
labeled_df.shape
```

```python
print("linhas nulas:")
print(labeled_df[labeled_df['labels'].isnull()])
```

```python
notNull_df = labeled_df[labeled_df['labels'].notnull()]
print(f"\ndf após remover linhas sem rótulos: {notNull_df.shape}")
```

```python
notNull_df = notNull_df.rename(columns={'text': 'texto', 'labels': 'rotulo', 'justification': 'argumento'})
df = notNull_df[['texto', 'rotulo', 'argumento']]
df.head()
```

```python
Path("datasets").mkdir(parents=True, exist_ok=True)
df.to_csv("datasets/base_df.csv", index=False, encoding="utf-8")

print("Arquivo salvo em: datasets/base_df.csv")
```
