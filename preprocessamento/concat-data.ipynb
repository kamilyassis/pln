```python
import pandas as pd
```

```python
df1 = pd.read_csv("https://media.githubusercontent.com/media/kamilyassis/pln/refs/heads/main/datasets/conexao-labels-full-4.1nano-p2.csv", sep="|")
print(df1.shape)
df1['source'] = 'conexao-labels-full-4.1nano-p2'
df1.head()
```

```python
df2 = pd.read_csv("https://media.githubusercontent.com/media/kamilyassis/pln/refs/heads/main/datasets/g1-labels-full-4.1nano-p2.csv")
print(df2.shape)
df2['source'] = 'g1-labels-full-4.1nano-p2'
df2.head()
```

```python
df3 = pd.read_csv("https://media.githubusercontent.com/media/kamilyassis/pln/refs/heads/main/datasets/g1_faketrue_treated.csv", sep="|")
print(df3.shape)
df3['source'] = 'g1_faketrue_treated'
df3.head()
```

```python
df_final = pd.concat([df1, df2, df3], ignore_index=True)
print(df_final.shape)
df_final.head()
```

```python
save = True
if save:
    df_final.to_csv("../datasets/combined_dataset.csv", index=False)
```
