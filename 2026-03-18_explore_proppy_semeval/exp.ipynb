```python
import pandas as pd
```

```python
df1 = pd.read_csv("data/semeval-subtask-3/dev-labels-subtask-3.txt", sep="\t", header=None, names=["article_id", "text_id", "label"])
print(df1.shape)
df1.head(10)
```

```python
df2 = pd.read_csv("data/semeval-subtask-3/dev-labels-subtask-3.template", sep="\t", header=None, names=["article_id", "text_id", "text"])
print(df2.shape)
df2.head(10)
```

```python
df = pd.merge(df1, df2, on=["article_id", "text_id"])
print(df.shape)
df.head(10)
```

```python
df1 = pd.read_csv("data/semeval-subtask-3/train-labels-subtask-3.txt", sep="\t", header=None, names=["article_id", "text_id", "label"])
print(df1.shape)
#print(df1.head())

df2 = pd.read_csv("data/semeval-subtask-3/train-labels-subtask-3.template", sep="\t", header=None, names=["article_id", "text_id", "text"])
print(df2.shape)
#print(df2.head())

_ = pd.merge(df1, df2, on=["article_id", "text_id"])
print(_.shape)
#print(_.head())

df = pd.concat([df, _], ignore_index=True)
print(df.shape)
print(df.head())
```
